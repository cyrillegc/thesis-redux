```python
import pandas as pd
import numpy as np
import re
```

```python
with open('data/data_concerts.csv', encoding='utf-8') as data_concerts:
    concerts_df = pd.read_csv(data_concerts, index_col=0)
```

```python
concerts_df
```

```python
artists_count = 0
artists_dict = dict()

for idx, row in concerts_df.iterrows():
    artist_id = 1

    while artist_id < 69:
        artist_name = row['artist_' + str(artist_id)]
        artist_url = row['artist_url_' + str(artist_id)]
        
        # keep strings in DataFrame
        if type(artist_name) == str:
            artists_dict[artist_url] = artist_name
            artists_count += 1
            
        # drop NaN values in DataFrame
        else:
            break

        artist_id += 1
        
print(artists_count)
print(len(artists_dict))
```

```python
# clean undesired indicators at the end of artist name, for example (CH) or (official)

for idx, artist_name in artists_dict.items():
    new_artist_name = re.sub(r'\([a-zA-Z]*\)$', '', artist_name)
    if new_artist_name:
        artists_dict[idx] = new_artist_name
        
len(artists_dict)
```

```python
artists_df = pd.DataFrame.from_dict(artists_dict, orient='index', columns=['artist_name']).sort_values(by=['artist_name'])
artists_df
```

```python
artists_df.to_csv('data/data_artists.csv', encoding='utf-8')
```
